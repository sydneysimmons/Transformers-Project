```python
from newspaper import Article
url = "https://www.cbssports.com/nfl/gametracker/recap/NFL_20220213_LAR@CIN/"
 
# download and parse article
article = Article(url)
article.download()
article.parse()
 
# print article text
text_3 = article.text
```

```python
words = text_3.split()

if len(words) > 512:
  words = words[:512]

text_3 = ' '.join(words)

summarizer_2(text_3)
```
